## Weather Data Visualization

```python
df = pd.read_csv("data\weather\synop.200701.csv",
            sep = ";")
```

```python
pd.to_datetime(df['date'], format='%Y%m%d%H%M%S')
```

```python
files_weather = os.listdir('data\weather')
files_map = {"2007": [], "2008": [], "2009": [], "2010": []}

for file in files_weather:
    if not file.endswith('.csv'):
        continue
    
    df = pd.read_csv(f'data\weather\{file}',
                     sep = ';')
    if "2007" in file:
        files_map["2007"].append(df)
    if "2008" in file:
        files_map["2008"].append(df)
    if "2009" in file:
        files_map["2009"].append(df)
    if "2010" in file:
        files_map["2010"].append(df)
```

```python
for k in files_map.keys():
    if isinstance(files_map[k], pd.DataFrame):
        continue
    files_map[k] = pd.concat(files_map[k])
```

```python
files_map['2007']
```

```python
df_stations = pd.read_csv("data\weather\postesSynop.csv",
            sep = ";")

for k in files_map.keys():
    files_map[k] = pd.merge(
        files_map[k],
        df_stations,
        left_on='numer_sta',
        right_on='ID',
        how= 'left'
    )
```

```python
files_map['2010']
```

```python
for k in files_map.keys():
    df = files_map[k]
    df.date = pd.to_datetime(df.date, format='%Y%m%d%H%M%S')
    files_map[k] = df
```

```python
# columns to be removed
for k in files_map.keys():
    files_map[k] = files_map[k].drop(columns=['Unnamed: 59','ID'], axis=1)
```

```python
# no rows to be removed due to na
files_map['2010'].isna().sum().sort_values(ascending = False)
```

```python
# rows to be removed
for k in files_map.keys():
    df = files_map[k]
    files_map[k] = df[df.numer_sta == 7149]
```

```python
def remove_outliers(df, columns, threshold=3):
    for col in columns:
        mu = df[col].mean()
        std = df[col].std()
        df['z_scores'] = np.abs((df[col] - mu) / std)
        df = df[df.z_scores < threshold]
        df.drop(columns=['z_scores'])

# Specify columns to detect outliers
columns_to_check = ['global_active_power']
# , 'global_reactive_power', 'voltage', 'global_intensity', 'sub_metering_1', 'sub_metering_2', 'sub_metering_3']

# Detect outliers using z-score method
remove_outliers(df_x, columns_to_check)
df_x
```
